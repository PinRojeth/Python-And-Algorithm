```python
# Algorithms 1

from heapq import heappush, heappop

def solve(cmd, heap) : 
    if cmd[0] == "push" : 
        heappush(heap, int(cmd[1]))
    elif cmd[0] == "pop" : 
        print(0 if not heap else heappop(heap))
        
n = int(input())
heap = []
for _ in range(n) : 
    cmd = input().split()
    solve(cmd, heap)
```

```python
from heapq import heappush, heappop

def solve(cmd, heap) : 
    if cmd[0] == "push" : 
        x = int(cmd[1])
        heappush(heap, (abs(x), x))
    elif cmd[0] == "pop" : 
        print(0 if not heap else heappop(heap)[1])

n = int(input())
heap = []
for _ in range(n) : 
    cmd = input().split()
    solve(cmd, heap)
```
