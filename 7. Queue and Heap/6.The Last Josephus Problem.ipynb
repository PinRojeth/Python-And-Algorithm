```python
from collections import deque 

def solve(n, k) : 
    queue = deque([i for i in range(1, n + 1)])
    while len(queue) > 1 : 
        for j in range(1, k + 1) : 
            if j != k : 
                queue.append(queue.popleft())
            else : 
                queue.popleft()
    return queue[0]
n, k = map(int, input().split())
print(solve(n, k))
```

```python
def solve(n, k) : 
    queue = [i for i in range(1, n + 1)]
    j = 0 
    
    while len(queue) > 1 : 
        j = (j + k - 1) % len(queue)
        queue.pop(j)
    return queue[0]


n, k = map(int, input().split())
import time 
start = time.time()
print(solve(n, k))
end = time.time()
print(f'solve() elapsed time : {end - start}')
```

```python
import sys 
sys.setrecursionlimit(10**7)

def solve(n, k) : 
    if n == 1 : 
        return 1
    else : 
        return ((solve(n - 1, k) + k - 1) % n) + 1
    
n, k = map(int, input().split())
import time 
start = time.time()
print(solve(n, k))
end = time.time()
print(f'solve() elapsed time : {end - start}')
```
