```python
# Solution 1

def binsearch(low, high, x, s) : 
    if low > high : 
        return -1
    else :
        mid = (high + low) // 2
        if s[mid] == x : 
            return mid
        elif s[mid] > x : 
            return binsearch(low, mid - 1, x, s)
        else : 
            return binsearch(mid + 1, high, x, s)
    
def solve(n, m, s, x) :
        for i in range(m) : 
            j = binsearch(0, n - 1, x[i], s)
            print(j, end=" ")
        print()
        
n, m = map(int, input().split())
s = list(map(int, input().split()))
x = list(map(int, input().split()))
solve(n, m, s, x)
```

```python
def binsearch(x, n, s) : 
    low = 0 
    high = n - 1
    while low <= high : 
        mid = (low + high) // 2 
        if x == s[mid] : 
            return mid 
        elif x < s[mid] : 
            high = mid - 1 
        else : 
            low = mid + 1
    return -1

def solve(n, m, s, x) :
        for i in range(m) : 
            j = binsearch(x[i], n, s)
            print(j, end=" ")
        print()
        
n, m = map(int, input().split())
s = list(map(int, input().split()))
x = list(map(int, input().split()))
solve(n, m, s, x)


# This is search x in s and it will show the index 
```
