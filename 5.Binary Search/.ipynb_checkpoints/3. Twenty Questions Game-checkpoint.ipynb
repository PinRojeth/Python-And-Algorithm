```python
def solve(x, s, t) : 
    low, high = s, t 
    while low < high : 
        mid = (low + high) // 2
        print("Bigger than mid?")
        if x > mid : 
            print("Yes.")
            low = mid - 1
        else : 
            print("No.")
            high = mid
        print("Your X is low.")
        return low
```
