```python
def cut(n, s, h) : 
    length = 0 
    for i in range(n) :
        if s[i] > h : 
            length += (s[i] - h)
    return length
 
def binsearch(low, high, n, m, s) : 
#     print(low, high)
    if low >= high : 
        return low - 1 
    else : 
        mid = (low + high) // 2 
        if cut(n, s, mid) < m : 
            result = binsearch(low, mid, n, m, s)
            return result
        else : 
            result = binsearch(mid + 1, high, n, m, s)
            return result
        
def solve(n, m, s)  : 
    j = binsearch(0, max(s), n, m, s) 
    print(j)
    
n, m = map(int, input().split())
s = list(map(int, input().split()))
solve(n, m, s)
```

```python
def cut(n, s, h) : 
    length = 0 
    for i in range(n) :
        if s[i] > h : 
            length += (s[i] - h)
    return length
 

def binsearch(n, m, s) : 
    low, high = 0, max(s)
    while low < high : 
        mid = (low + high) // 2 
        if cut(n, s, mid) < m : 
            high = mid 
        else : 
            low = mid + 1
    return low -1 

def solve(n, m, s)  : 
    j = binsearch(n, m, s) 
    print(j)
    
n, m = map(int, input().split())
s = list(map(int, input().split()))
solve(n, m, s)
```
