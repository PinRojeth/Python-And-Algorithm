```python
# Solution 1 using Iteration

def solve(x, s, t) : 
    low, high = s, t 
    while low < high : 
        mid = (low + high) // 2
        print(f"Bigger than {mid}?", end=" ")
        if x > mid : 
            print("Yes.")
            low = mid + 1
        else : 
            print("No.")
            high = mid
    print(f"Your X is {low}.")
    return low
    
import random 
s, t = map(int, input().split())
SEED = int(input())
random.seed(SEED)
x = random.randint(s, t)
solve(x, s, t)

```

```python
def solve(x, low, high) : 
    if low >= high : 
        print(f"Your X is {low}")
        return low 
    else :
        mid = (low + high) // 2 
        print(f"Bigger than {mid}", end=" ")
        if x > mid : 
            print('Yes')
            return solve(x, mid + 1, high)
        else : 
            print("No")
            return solve(x, low, mid)

import random 
s, t = map(int, input().split())
SEED = int(input())
random.seed(SEED)
x = random.randint(s, t)
solve(x, s, t)
```
