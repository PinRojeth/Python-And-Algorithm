```python
def seqsearch(x, n, s) :
    for i in range(n) :
        if x == s[i] :
            return i
    return -1

def solve(n, m, s, x) :
    for i in range(m) : 
        print(seqsearch(x[i], n, s), end = " " )
    print()
    
n, m = map(int, input().split())
s = list(map(int, input().split()))
x = list(map(int, input().split()))
solve(n, m, s, x)
```
