```python
def is_prime(n) :
    if n <= 1 :
        return False
    for i in range(2, int(n**0.5) + 1) : 
        if n % i == 0 : 
            return False
    return True

def solve(n) :
    cnt = 0 
    for i in range(2, n + 1) :
        if is_prime(i) :
            cnt += 1 
    return cnt 

n = int(input())
solve(n)
```

```python
def find_primes(n) :
    sieve = [0, 0] + [1] * (n - 1)
    for i in range(2, n + 1) :
        if sieve[i] == 1 : 
            for j in range(i + i, n + 1, i):
                sieve[j] = 0 
    return sieve

def solve(n) :
    sieve = find_primes(n)
    return sum(sieve)
n = int(input())
print(solve(n))
```
